```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("breast-cancer.csv")
data.head()
```

```python
data = data.drop('id',axis=1)
data
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
data['diagnosis'] = encoder.fit_transform(data['diagnosis'])
data
```

```python
plt.figure(figsize=(30,45))
sns.heatmap(data.corr(),cmap="Greens",annot=True)
plt.plot()
```

```python
data['diagnosis']
```

```python
x = data.drop('diagnosis',axis=1)
y = data['diagnosis']
```

```python
sns.countplot(x='radius_mean',hue='diagnosis',data=data)
plt.plot()
```

```python
x.head()
y.head()
```

```python
x = x.drop(columns=['fractal_dimension_mean','texture_se','smoothness_se','symmetry_se','fractal_dimension_se'],axis=1)
# data
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.svm import SVC

pred = SVC(kernel='linear',random_state=0)
pred.fit(x_train,y_train)
```

```python
y_pred = pred.predict(x_test)
```

```python
from sklearn import metrics

cm = metrics.confusion_matrix(y_pred,y_test)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = [True,False])
cm_display.plot()
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test))
```
